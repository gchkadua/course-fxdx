ერთ–ერთი მთავარი მიზეზი რის გამოც ვანვითარებთ კომპიუტერულ პროგრამებს არის ის რომ მოვახდინოთ სხვადასხვა ტიპის პროცესების ავტომატიზაცია. კერძოდ მანქანური დასწავლა წარმოადგენს ხელოვნური ინტელექტის ქვე მიმართულებას რომლის ზოგადი მიზანიც არის კომპიუტერული პრგრამების "ხელით ჩარევის" გარეშე განვითარება. ასე რომ უხეშად შეგვიძლია ვთქვათ რომ მანქანური დასწავლა არის "ავტომატიზაციის ავტომატიზაცია". სხვა სიტყვებით მანქანური დასწავლა აძლევს კომპიუტერს საშუალებას შექმნას პროგრამები თავად. შეგვიძლია ვთქვათ რომ  ML არის მონაცემების პროგრამად გარდაქმნის პროცესი.

მანქანური დასწავლა შეიძლება იყოს supervised, unsupervised, semi-supervised  და reinforcement. ჩვენ კურსის ფარგლებში შევეხებით ძირიტად პირველ ორ ტიპს.

ჩვენს წინა ლექციაში განხილული მონაცემები იყო  supervised მანქანური დასწავლის შესაბამისი მაგალითი, ანუ სადაც მონაცემებს გააჩნიათ label–ები, (ანუ ცლვადების სახელები) და ზუსტად ვიცით საპროგნოზო ცვლადის სახელიც. 

supervised ამოცანა ასევე იყოფა ორი ტიპის ამოცანად

1. კლასიფიკაცია

ამ შემთხვევაში ჩვენ გვაქვს ორი კლასი და ამოცანა არის ისე ვასწავლოთ "მოდელს", რომ მას შემდგომ შეეძლოს  ამ ორი კლასის განსხვავება გარკვეული სიზუსტით

2.რეგრესია

ეხლა ჩვენ გვაქვს მოყვანილი უმარტივესი ერთი ცვლადის (x) წრფივი რეგრესიის მოდელი საპროგნოზო ცვლადი y-ით.
ამოცანა კი არის შემდეგი რომ გავავლოთ ისეთი წრფე (მოდელი) რომ მომავალში ახლად შემოსული მონაცემები საშუალოდ მინიმალურად იყვნენ დაშორებული ამ წრფიდან.

ფაქტიურად ეს ორი იდეა არის მანქანრუი დასწავლის (და თამამად შეგვიძლია ვთქვათ ხელოვნური ინტელექტის ფუნდამენტი), თუ კი ეს კონცეფცია გვესმის კარგად შეგვიძლია ჩავთვალოთ რომ ცვენ მანქანური დასწავლის და ხელოვნური ინტელექტის უზარაზარი სფეროს კარები შეღებული გვაქვს.

შემდეგი კატეგორიის ამოცანა არის unsupervised  მაქნანური დასწავლა, რაც გულისხმობს რომ ჩვენ საქმე გვაქვს მონაცემებთან რომელთაც label–ები არ გააჩნიათ. ყველაზე გავრცელებული unsupervised მანქანური დასწავლის ამოცანა არის კლასტერების შექმნა:

ჩვენ დღეს შევეხებით რეგრესიის და კლასიფიკაციის ამოცანების ტრივიალურ მაგალითებს.

supervised ტიპის ამოცანებს გააჩნიათ overfitting–ის საფრთხე. რაც იმას ნიშნავს რომ მოდელი კარგად პრპგნოზირებს საწვრთნელ მონაცემებზე მაგრამ სატესტო, "უნახავ" მონაცემებზე აქვს შედარებით ცუდი შედეგი. რაც უფრო კომპლექსური ხდება მოდელი მით უფრო იზრდება შანსი დადეს overfitting–ის საფრთხე.  ეს მოვლენა მატემატიკურად ნიშნავს მონაცემებში მარალი ვარიანტის ქონას.  

სწორედ მსგავსი საფრთხეების თავიდან ასარიდებლად გამოყენება მონაცემის 3 ნაწილად გაყოფა
1. საწვრთნელი ნაწილი– სადაც ჩვენი მოდელი სწავლობს კანონზომიერებებს
2. ვალიდაციის ნაწილი – სადაც ვამოწმებთ თუ რამდენად კარგად ისწავლა ჩვენმა მოდელმა 
3. სატესტო ნაწილი– სადაც ვცდით თუ როგორი წარმადობა ექნება მოდელს მომავალში სრულიად უნახავ მონაცემებზე

მაგ. თუ მონაცემების წარმადობა საწვრთნელ და ვალიდაციის ნაწილებზე ერთნაირად კარგია (ანუ არ გვიდგას overfitting-ის პრობლემა) ჩვენ გვაქვს უფლება სატესტო ნაწილზე შევამოწმოთ მოდელის წარმადობა. მიღებული შედეგის საფუძველზე შევძლებთ დასკვნა გავაკეთოთ მოდელის ავკარგიანობაზე

```python
dataset = pd.read_csv('User_Data.csv')
```

```python
dataset.shape
```

```python
dataset.fillna(method ='ffill', inplace = True)
```

```python
#მოვახდინოთ ენკოდირება
# from sklearn.preprocessing import LabelEncoder, OneHotEncoder
# le = LabelEncoder()
# dataset['Gender'] = le.fit_transform(dataset['Gender'].astype(str))
```

```python
X = dataset.drop(columns=['User ID', 'Gender', 'Purchased'])
y = dataset['Purchased']
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0, stratify=y)
```

```python
#სკალირება და სტანდარტიზაცია
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train) 
x_test = sc.transform(x_test)
```

```python
print (x_train[0:10, :])
```

```python
#შემოვიტანოთ ლოჯისტიკური რეგრესიის ბიბლიოთეკა და განვმარტოთ მოდელი
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
#საწვრთნელ მონაცმეებზე მოვარგოთ მოდელი
classifier.fit(x_train, y_train)
```

```python
#ვიპროგონოზოთ სატესტო მონაცემებზე
y_pred = classifier.predict(x_test)
```

```python
#ვნახოთ რა სიზუსტით გამოვიცანით
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score
print ("Accuracy : ", accuracy_score(y_test, y_pred))
```

```python
from matplotlib.colors import ListedColormap
X_set, y_set = x_test, y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, 
                               stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, 
                               stop = X_set[:, 1].max() + 1, step = 0.01))
  
plt.contourf(X1, X2, classifier.predict(
             np.array([X1.ravel(), X2.ravel()]).T).reshape(
             X1.shape), alpha = 0.75, cmap = ListedColormap(('red', 'green')))
  
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
  
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('blue', 'white'))(i), label = j)
      
plt.title('Classifier (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
